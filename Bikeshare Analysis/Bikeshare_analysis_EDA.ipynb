# Feature Engineering

```python
# Creating new columns from datetime columns
```

```python
type(dataset['datetime'].iloc[0])
```

```python
# Changing the format of datetime column from string to Datetime objects
dataset['datetime'] = pd.to_datetime(dataset['datetime'])
```

```python
type(dataset['datetime'].iloc[0])
```

Create new columns "date,"hour","weekDay","month" from "datetime" column.

```python
dataset['date'] = dataset['datetime'].iloc[0]
dataset['hour'] = dataset['datetime'].apply(lambda time: time.hour)
dataset['month'] = dataset['datetime'].apply(lambda time: time.month)
dataset['weekday'] = dataset['datetime'].apply(lambda time: time.dayofweek)
```

```python
dataset.head(2)
```

```python
type(dataset['date'].iloc[0])
```

Mapping values to categories

```python
dataset["season"] = dataset.season.map({1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" })
dataset["weather"] = dataset.weather.map({1: " Clear + Few clouds + Partly cloudy + Partly cloudy",\
                                        2 : " Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist ", \
                                        3 : " Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds", \
                                        4 :" Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog " })
```

Creating a category type list

```python
categoryVariableList = ["hour","weekday","month","season","weather","holiday","workingday"]
for variable in categoryVariableList:
    dataset[variable] = dataset[variable].astype("category")
```

```python
dataset.head(2)
```

Dropping the datetime column

```python
dataset = dataset.drop(["datetime"],axis=1)
```

```python
dataset.head(2)
```

# Outlier Analysis

Creating boxplots to check for outliers in the count column

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,10))

sns.boxplot(data=dataset, y="count",orient="v",ax=axes[0][0])
sns.boxplot(data=dataset, x="season", y="count",orient="v",ax=axes[0][1])
sns.boxplot(data=dataset, x="hour", y="count",orient="v",ax=axes[1][0])
sns.boxplot(data=dataset, x="workingday", y="count",orient="v",ax=axes[1][1])

axes[0][0].set(ylabel='Count',title="Box Plot On Count")
axes[0][1].set(xlabel='Season', ylabel='Count',title="Box Plot On Count Across Season")
axes[1][0].set(xlabel='Hour Of The Day', ylabel='Count',title="Box Plot On Count Across Hour Of The Day")
axes[1][1].set(xlabel='Working Day', ylabel='Count',title="Box Plot On Count Across Working Day")
plt.tight_layout()
```

# From the above boxplots the following inferences can be made:

<ol>
<li> Spring season has a very low count as there is a clear drop in the median value.</li> 
<li>During the hours between 7-8 and 17-18 we can clearly see that median values are relatively higher.</li> 
<li>Outliers are mostly on the working days rather than non working days.</li> 
</ol>

<H2> Removing Outliers </H2>

```python
dataset_without_outliers = dataset[np.abs(dataset["count"]-dataset["count"].mean()) <= (3*dataset["count"].std())]
```

```python
print ("Shape Of The Before Ouliers: ",dataset.shape)
print ("Shape Of The After Ouliers: ",dataset_without_outliers.shape)
```

# Correlation
<h3>Checking correlation between numerical features</h3>

```python
dataset_without_outliers.head(2)
```

```python
correlation_matrix = dataset_without_outliers[["temp","atemp","humidity","windspeed","casual","registered","count"]].corr()
correlation_matrix
mask = np.array(correlation_matrix)
mask[np.tril_indices_from(mask)] = False
```

```python
fig,ax= plt.subplots(figsize=(20,10))
sns.heatmap(correlation_matrix, mask=mask,vmax=.8, square=True,annot=True)
plt.tight_layout()
```

RegPlots

```python
fig,(ax1,ax2,ax3) = plt.subplots(ncols=3, figsize=(12,6))
sns.regplot(x="temp",y="count",data=dataset_without_outliers, ax=ax1)
sns.regplot(x="windspeed",y="count",data=dataset_without_outliers, ax=ax2)
sns.regplot(x="humidity",y="count",data=dataset_without_outliers, ax=ax3)
plt.tight_layout()
```

<h3> Inferences </h3>
<ol>
    <li>temp has a positive correlation with count, There is little dependency</li>
    <li>humidity has a negative correlation with count, There is little dependency</li>
    <li>windspeed has a very low correlation value with count, Thus very little dependency</li>
    <li>temp and atemp has strong correlation, Thus one of them should be dropped during model building to avoid multicollinearity  in the data.</li>
</ol>

```python
fig, axes = plt.subplots(ncols = 2, nrows = 1,figsize=(12, 6))
sns.distplot(dataset["count"],ax=axes[0])
sns.distplot(dataset_without_outliers["count"],ax=axes[1])
plt.tight_layout()
```

Here, we can see that there is a possible skewness in the count. Taking log of the count to make the distribution normal.

```python
fig, axes = plt.subplots(ncols = 2, nrows = 1,figsize=(10, 6))
sns.distplot(dataset_without_outliers["count"],ax=axes[0])
sns.distplot(np.log(dataset_without_outliers["count"]),ax=axes[1])
plt.tight_layout()
```

Not an ideal normal distribution but better than before.

```python
fig, axes = plt.subplots(figsize=(8,4), dpi=100)
monthcluster = pd.DataFrame(dataset_without_outliers.groupby("month")["count"].mean()).reset_index()
monthsorted = monthcluster.sort_values(by="count",ascending=False)
sns.barplot(data=monthsorted,x="month",y="count")
axes.set(xlabel='Month', ylabel='Avearage Count',title="Average Count per Month")
plt.tight_layout()
```

```python
# Mapping actual strings to Day of Week
dataset_with_weekday = dataset_without_outliers.copy()
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
dataset_with_weekday['weekday'] = dataset_with_weekday['weekday'].map(dmap)
```

```python
dataset_with_weekday.head()
```

```python
fig, axes = plt.subplots(figsize=(12,4), dpi=100)
hourcluster = pd.DataFrame(dataset_with_weekday.groupby(["hour","weekday"],sort=False)["count"].mean()).reset_index()
sns.pointplot(x=hourcluster["hour"], y=hourcluster["count"],hue=hourcluster["weekday"],data=hourcluster, join=True)
axes.set(xlabel='Hour Of The Day', ylabel='Users Count',title="Average Users Count By Hour Of The Day Across Weekdays")
axes.legend(loc=0)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(12,4), dpi=100)
hourcluster = pd.DataFrame(dataset_without_outliers.groupby(["hour","season"],sort=False)["count"].mean()).reset_index()
sns.pointplot(x=hourcluster["hour"], y=hourcluster["count"],hue=hourcluster["season"],data=hourcluster, join=True)
axes.set(xlabel='Hour Of The Day', ylabel='Users Count',title="Average Users Count By Hour Of The Day Across Season")
axes.legend(loc=0)
plt.tight_layout()
```

```python
transformed_user_type = pd.melt(dataset_without_outliers[["hour","registered","casual"]],id_vars=["hour"],value_vars=["registered","casual"])
```

```python
fig, axes = plt.subplots(figsize=(12,4), dpi=100)
hourcluster = pd.DataFrame(transformed_user_type.groupby(["hour","variable"],sort=True)["value"].mean()).reset_index()
sns.pointplot(x=hourcluster["hour"], y=hourcluster["value"],hue=hourcluster["variable"], data=hourcluster, join=True)
axes.set(xlabel='Hour Of The Day', ylabel='Users Count',title="Average Users Count By Hour Of The Day Across User Type")
axes.legend(loc=0)
plt.tight_layout()
```

<h3> Inferences </h3>
<ol>
    <li>People tend to rent bike during summer season since it is really conducive to ride bike at that season.Therefore months 6,7,8 has got relatively higher demand for bicycle.</li>
    <li>On weekdays more people tend to rent bicycle around 7AM-8AM and 5PM-6PM.</li>
    <li>On weekends People tend to rent between 10AM and 4PM.</li>
    <li>The peak user count around 7AM-8AM and 5PM-6PM is purely contributed by registered user.</li>
</ol>
